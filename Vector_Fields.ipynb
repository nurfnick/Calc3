```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

M = 2*x*y
N = x*x+6*y*y

plt.quiver(x,y,M,N)
plt.show()
```
